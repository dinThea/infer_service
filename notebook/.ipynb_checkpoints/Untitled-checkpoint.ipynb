```python
!pip3 install matplotlib
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
def getOutputLayers(net):
    ln = net.getLayerNames()
    ln = [ln[i[0] - 1] for i in net.getUnconnectedOutLayers()]
    
    return ln
```

```python
net = cv2.dnn.readNetFromDarknet('yolo.cfg', 'yolov3.weights')
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_DEFAULT)
```

```python
print (os.listdir('.'))
```

```python
img = cv2.imread('How-to-Grow-Broccoli-at-Home.jpg')
```

```python
img2 = img[:,:,::-1]
plt.imshow(img2)
print (img2.shape)
```

```python
blob = cv2.dnn.blobFromImage(img, 1.0/255, (416, 416), swapRB=True, crop=False)
```

```python
net.setInput(blob)
outs = net.forward(getOutputLayers(net))
```

```python
objects = []
threshold = 0.5

for out in outs:
    
    for detection in out:
        scores = detection[5:]
        max_score = np.argmax(scores)
        score = scores[max_score]
        print (score)
```
