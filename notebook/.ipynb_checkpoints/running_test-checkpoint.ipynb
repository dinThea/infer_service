```python
%%capture
%matplotlib inline
!pip3 install matplotlib
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize = (5,5))
```

```python
def getOutputLayers(net):
    ln = net.getLayerNames()
    ln = [ln[i[0] - 1] for i in net.getUnconnectedOutLayers()]
    
    return ln
```

```python
net = cv2.dnn.readNetFromDarknet('yolo.cfg', 'yolov3.weights')
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_DEFAULT)
```

```python
print (os.listdir('.'))
```

```python
img = cv2.imread('How-to-Grow-Broccoli-at-Home.jpg')
```

```python
img2 = img[:,:,::-1]
plt.imshow(img2)
```

```python
blob = cv2.dnn.blobFromImage(img, 1.0/255, (416, 416), swapRB=True, crop=False)
```

```python
net.setInput(blob)
outs = net.forward(getOutputLayers(net))
```

```python
objects = []
threshold = 0.5
h, w = img.shape[:2]
print (h, w)
for out in outs:
    
    for detection in out:
        scores = detection[5:]
        max_score = np.argmax(scores)
        score = scores[max_score]
        if score >= threshold:
            centerx, centery, width, height = tuple(int(val*multiplier) for val, multiplier in zip(detection[0:4], [w,h,w,h]))
            x, y = int(centerx - width/2), int(centery - height/2)
            objects.append(((x,y),(x+width, y+height)))
            
```

```python
print (objects)
```

```python
res = img.copy()
for pos, size in objects:
    print (pos, size)
    cv2.rectangle(res, pos, size, (0,255,0), 3)
```

```python
res2 = res[:,:,::-1]
plt.imshow(res2)
```
